```python
#import necessary tools
import re
from string import digits
import pandas as pd
```

```python
#read the documents
with open('102_long.txt','rb') as f:
    contents = f.read()
    contents = contents.decode("utf-16").rstrip("\n")
    contents = contents.split("\r\n") #maybe need to strip \t?
```

```python
BlockNum=[]
Procedure=list()
TrialNUM =[] #somehow eprime writes as TrialNUM
img1 =[]
img2=[]
ChoiceB_RESP = []
ChoiceB_RT=[]

for i in range(len(contents)):
    if contents[i] == '\t*** LogFrame Start ***' and contents[i+4].startswith("\tTrial")==True:
        Procedurea = contents[i+1].split("Procedure: ",1)[1] 
        Procedure.append(Procedurea)
        
        
        img1a =contents[i+2].split("img1: ",1)[1] 
        img1.append(img1a)
        
        img2a =contents[i+3].split("img2: ",1)[1] 
        img2.append(img2a)
        
        TrialNUMa =[''.join(c for c in contents[i+4] if c in digits)]
        TrialNUM=TrialNUM+TrialNUMa
        
        BlockNuma =[''.join(c for c in contents[i+5] if c in digits)]
        BlockNum=BlockNum+BlockNuma
        
        ChoiceB_RESPa =[''.join(c for c in contents[i+15] if c in digits)]
        ChoiceB_RESP=ChoiceB_RESP+ChoiceB_RESPa
        
        ChoiceB_RTa =[''.join(c for c in contents[i+14] if c in digits)]
        ChoiceB_RT=ChoiceB_RT+ChoiceB_RTa
    
Procedure_df = pd.DataFrame(Procedure)
Procedure_df.columns = ['Procedure']  
        
img1_df =pd.DataFrame(img1)
img1_df.columns = ['img1']

img2_df =pd.DataFrame(img2)
img2_df.columns = ['img2']

TrialNUM_df =pd.DataFrame(TrialNUM)
TrialNUM_df.columns = ['TrialNUM']

BlockNum_df =pd.DataFrame(BlockNum)
BlockNum_df.columns = ['BlockNum']

ChoiceB_RESP_df =pd.DataFrame(ChoiceB_RESP)
ChoiceB_RESP_df.columns = ['ChoiceB_RESP']

ChoiceB_RT_df =pd.DataFrame(ChoiceB_RT)
ChoiceB_RT_df.columns = ['ChoiceB_RT']

all = pd.concat([img1_df.reset_index(drop=True), img2_df], axis=1)
all = pd.concat([all.reset_index(drop=True), Procedure_df], axis=1)
all = pd.concat([all.reset_index(drop=True), TrialNUM_df], axis=1)
all = pd.concat([all.reset_index(drop=True), BlockNum_df], axis=1)
all = pd.concat([all.reset_index(drop=True), ChoiceB_RESP_df], axis=1)
all = pd.concat([all.reset_index(drop=True), ChoiceB_RT_df], axis=1)
```
