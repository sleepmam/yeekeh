```python
#import necessary tools
import re
from string import digits
import pandas as pd
import numpy as np
```

```python
#read the documents
with open('209_1.txt','rb') as f:
    contents = f.read()
    contents = contents.decode("utf-16").rstrip("\n")
    contents = contents.split("\r\n") #maybe need to strip \t?
```

```python
BlockNum=[]
Procedure=list()
TrialNUM =[] #somehow eprime writes as TrialNUM
img1 =[]
img2=[]
Choice_RESP = []
Choice_RT=[]
#there should be a BL and LB version
#101 is BL, 16 RT, 


#first determine if this is our target block, with  logframe start, +8 is list something
#second, common theme such as img1, img2, 
#third identify specific group, specficy a new index, and set index that way?
#four,  fit into each group with the given index. 

for i in range(len(contents)):
    
    if contents[i] == '\t*** LogFrame Start ***' and contents[i+8].startswith("\tListBlack1") == True:
        #black 1
        Procedurea = contents[i+1].split("Procedure: ",1)[1] 
        Procedure.append(Procedurea)        
        
        img1a = contents[i+2].split("img1: ",1)[1] 
        img1.append(img1a)
        
        img2a = contents[i+3].split("img2: ",1)[1] 
        img2.append(img2a)
        
        TrialNUMa =[''.join(c for c in contents[i+4] if c in digits)]
        TrialNUMa = [float(i) for i in TrialNUMa] 
        TrialNUMa = list(TrialNUMa)
        TrialNUM=TrialNUM+TrialNUMa
        
        BlockNuma =[''.join(c for c in contents[i+5] if c in digits)]
        BlockNum=BlockNum+BlockNuma
        
        Choice_RESPa =[''.join(c for c in contents[i+17] if c in digits)]
        Choice_RESP=Choice_RESP+Choice_RESPa
        
        Choice_RTa =[''.join(c for c in contents[i+16] if c in digits)]
        Choice_RTa = [float(i) for i in Choice_RTa]
        Choice_RTa = list(Choice_RTa)
        Choice_RT=Choice_RT+Choice_RTa
        
    elif contents[i] == '\t*** LogFrame Start ***' and contents[i+8].startswith("\tListBlack2") == True:
        #black 2
        Procedurea = contents[i+1].split("Procedure: ",1)[1] 
        Procedure.append(Procedurea)        
        
        img1a =contents[i+2].split("img1: ",1)[1] 
        img1.append(img1a)
        
        img2a =contents[i+3].split("img2: ",1)[1] 
        img2.append(img2a)
        
        TrialNUMa =[''.join(c for c in contents[i+4] if c in digits)]
        TrialNUMa = [float(i) for i in TrialNUMa] 
        TrialNUMa = list(TrialNUMa)
        TrialNUM=TrialNUM+TrialNUMa
        
        BlockNuma =[''.join(c for c in contents[i+5] if c in digits)]
        BlockNum=BlockNum+BlockNuma
        
        Choice_RESPa =[''.join(c for c in contents[i+15] if c in digits)]
        Choice_RESP=Choice_RESP+Choice_RESPa
        
        Choice_RTa =[''.join(c for c in contents[i+14] if c in digits)]
        print(Choice_RTa)
        if Choice_RTa[-1] == '':
            continue
        Choice_RTa = [float(i) for i in Choice_RTa]
        Choice_RTa = list(Choice_RTa)
        Choice_RT=Choice_RT+Choice_RTa
        
    elif contents[i] == '\t*** LogFrame Start ***' and contents[i+8].startswith("\tListLike1") == True:
        #like 1
        Procedurea = contents[i+1].split("Procedure: ",1)[1] 
        Procedure.append(Procedurea)        
        
        img1a =contents[i+2].split("img1: ",1)[1] 
        img1.append(img1a)
        
        img2a =contents[i+3].split("img2: ",1)[1] 
        img2.append(img2a)
        
        TrialNUMa =[''.join(c for c in contents[i+5] if c in digits)]
        TrialNUMa = [float(i) for i in TrialNUMa] 
        TrialNUMa = list(TrialNUMa)
        TrialNUM=TrialNUM+TrialNUMa
        
        BlockNuma =[''.join(c for c in contents[i+6] if c in digits)]
        BlockNum=BlockNum+BlockNuma
        
        Choice_RESPa =[''.join(c for c in contents[i+18] if c in digits)]
        Choice_RESP=Choice_RESP+Choice_RESPa
        
        Choice_RTa =[''.join(c for c in contents[i+17] if c in digits)]
        Choice_RTa = [float(i) for i in Choice_RTa]
        Choice_RTa = list(Choice_RTa)
        Choice_RT=Choice_RT+Choice_RTa
        
    elif contents[i] == '\t*** LogFrame Start ***' and contents[i+8].startswith("\tListLike2") == True:
        #like 2
        Procedurea = contents[i+1].split("Procedure: ",1)[1] 
        Procedure.append(Procedurea)        
        
        img1a =contents[i+2].split("img1: ",1)[1] 
        img1.append(img1a)
        
        img2a =contents[i+3].split("img2: ",1)[1] 
        img2.append(img2a)
        
        TrialNUMa =[''.join(c for c in contents[i+5] if c in digits)]
        TrialNUMa = [float(i) for i in TrialNUMa] 
        TrialNUMa = list(TrialNUMa)
        TrialNUM=TrialNUM+TrialNUMa
        #trial needs to be plus one, the NONE group is also not right yet
        #needs to be (blocknum-1)*50+trial
        
        BlockNuma =[''.join(c for c in contents[i+6] if c in digits)]
        BlockNum=BlockNum+BlockNuma
        
        Choice_RESPa =[''.join(c for c in contents[i+16] if c in digits)]
        Choice_RESP=Choice_RESP+Choice_RESPa
        
        Choice_RTa =[''.join(c for c in contents[i+15] if c in digits)]
        Choice_RTa = [float(i) for i in Choice_RTa]
        Choice_RTa = list(Choice_RTa)
        Choice_RT=Choice_RT+Choice_RTa
        
        
Procedure_df = pd.DataFrame(Procedure)
Procedure_df.columns = ['Btype']  
        
img1 = [int(i) for i in img1] 
img1_df = pd.DataFrame(img1)
img1_df.columns = ['img1']

img2 = [int(i) for i in img2] 
img2_df = pd.DataFrame(img2)
img2_df.columns = ['img2']

TrialNUM = [int(i) for i in TrialNUM] 
TrialNUM_df = pd.DataFrame(TrialNUM)
TrialNUM_df.columns = ['trial']

BlockNum = [int(i) for i in BlockNum] 
BlockNum_df = pd.DataFrame(BlockNum)
BlockNum_df.columns = ['block']

Choice_RESP = [int(i) for i in Choice_RESP] 
Choice_RESP_df = pd.DataFrame(Choice_RESP)
Choice_RESP_df.columns = ['RESP']

Choice_RT = [int(i) for i in Choice_RT] 
Choice_RT_df = pd.DataFrame(Choice_RT)
Choice_RT_df.columns = ['RT']
```

```python
sub_id = contents[16].split("Subject: ",1)[1] 
sub_df = pd.DataFrame(np.repeat(sub_id, img1_df.shape[0], axis = 0))
sub_df.columns = ['sub']

all = pd.concat([sub_df.reset_index(drop=True), BlockNum_df], axis=1)
all = pd.concat([all.reset_index(drop=True), Procedure_df], axis=1)
all = pd.concat([all.reset_index(drop=True), TrialNUM_df], axis=1)
all = pd.concat([all.reset_index(drop=True), img1_df], axis=1)
all = pd.concat([all.reset_index(drop=True), img2_df], axis=1)
all = pd.concat([all.reset_index(drop=True), Choice_RESP_df], axis=1)
all = pd.concat([all.reset_index(drop=True), Choice_RT_df], axis=1)

all
```
